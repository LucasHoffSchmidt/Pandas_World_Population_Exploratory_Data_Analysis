# Pandas project - Exploratory Data Analysis

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Transform a csv file to a dataframe
df = pd.read_csv(r"C:\Users\lucas\Desktop\Software_Development\Data Analytics\Python\Projects\Pandas - Exploratory data analysis\world_population.csv")
df
```

```python
# Format each number to two decimal places
pd.set_option("display.float_format", lambda x: "%.2f" % x)
```

```python
# Show columns non-null count and data type
df.info()
```

```python
# Show statistical elements for each column
df.describe()
```

```python
# Shows how many null values each column has
df.isnull().sum()
```

```python
# Shows number of unique values in each column
df.nunique()
```

```python
# Select columns by data type
df.select_dtypes(include="number")
```

```python
# Show the top 10 lowest values for the column 2022 population
df.sort_values(by="2022 Population").head(10)
```

```python
# Show the top 10 highest values for the column 2022 population
df.sort_values(by="2022 Population", ascending=False).head(10)
```

```python
# Sort dataframe descendingly by world population percentage
df.sort_values(by="World Population Percentage", ascending=False).head(10)
```

```python
# Show the numerical correlation between columns
df_numeric = df.select_dtypes(include="number")
df_numeric.corr()
```

```python
# Shows the correlation between columns visually using seaborn and matplotlib

# Makes a correlation heatmap, showing the correlation numbers
sns.heatmap(df_numeric.corr(), annot = True)

# Gives the heatmap a title and changes its size
plt.title("Correlation Matrix")
plt.rcParams["figure.figsize"] = (20, 7)
plt.show()
```

```python
# Shows the mean of each continent, sorted by the 2022 Population, descendingly
df.groupby("Continent").mean(numeric_only=True).sort_values(by="2022 Population", ascending=False)
```

```python
# Show all values of the Continent column that contains the string "Oceania"
df[df["Continent"].str.contains("Oceania")]
```

```python
# Shows a list of all column names
df.columns
```

```python
# Shows the mean of each continent, sorted by the 2022 Population descendingly and only including population data
df2 = df.groupby("Continent")[df.columns[5:13][::-1]].mean(numeric_only=True).sort_values(by="2022 Population", ascending=False)
df2
```

```python
# Transpose the axes to show continents in the columns and populations in the rows
df3 = df2.transpose()
df3
```

```python
# Show a line plot of the growth of continents
df3.plot()
```

```python
# Show outliers by a boxplot
df.boxplot(figsize=(20,10))
```
